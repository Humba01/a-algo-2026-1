# **Implementation**

```python
import datetime as dt
import numpy as np
import os
import random as rnd
import zoneinfo as zi

"""
Setting example constants.
"""
values = []
SZ_1ML = 1000
SZ_5ML = 5000
SZ_10ML = 10000
SZ_20ML = 20000
SZ_50ML = 50000

"""
Setting the zone info
"""
tz = zi.ZoneInfo("America/Sao_Paulo")

"""
Adding random values in variable "VALUES".
"""
def generate_numbers(size):
    for i in range(size):
        values.append(int(rnd.random() * 100))
    return values

"""
Function insertionsort (n²).
"""
def insertion_sort(num: int):
  values.clear()
  generate_numbers(num)
  print(f"\x1b[31;1mInsertion Sort (n²) - With {num} itens:\x1b[0m")
  start = dt.datetime.now(tz)
  print(f"\t\x1b[31;1m[Start]⇨ {start}\x1b[0m")

  for x in range(len(values)):
    key = values[x]
    y = x - 1

    while y >= 0 and key < values[y]:
      values[y + 1] = values[y]
      y -= 1
    values[y + 1] = key

  end = dt.datetime.now(tz)
  print(f"\t\x1b[31;1m[End]⇨ {end}\x1b[0m")
  print(f"\t\x1b[31;1m[Interval]⇨ {(end-start).microseconds}ms\x1b[0m")
  breakline()

"""
Function timsort (n log n).
"""
def tim_sort(num: int):
  values.clear()
  generate_numbers(num)
  print(f"\x1b[34;1mTim Sort (n log n) - With {num} itens:\x1b[0m")
  start = dt.datetime.now(tz)
  print(f"\t\x1b[34;1m[Start]⇨ {start}\x1b[0m")
  values.sort()
  end = dt.datetime.now(tz)
  print(f"\t\x1b[34;1m[End]⇨ {end}\x1b[0m")
  print(f"\t\x1b[34;1m[Interval]⇨ {(end-start).microseconds}ms\x1b[0m")
  breakline()

"""
Function breakline
"""
def breakline():
  print("\n")

"""
Calling functions to compare then.
"""
insertion_sort(SZ_1ML)
insertion_sort(SZ_5ML)
insertion_sort(SZ_10ML)
insertion_sort(SZ_20ML)
insertion_sort(SZ_50ML)
tim_sort(SZ_1ML)
tim_sort(SZ_5ML)
tim_sort(SZ_10ML)
tim_sort(SZ_20ML)
tim_sort(SZ_50ML)
breakline()
print("End of process.")
```
