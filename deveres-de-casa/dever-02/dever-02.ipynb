## **Implementation**

```python
import datetime as dt
import numpy as np
import os
import random as rnd
import zoneinfo as zi

"""
Setting the zone info
"""
tz = zi.ZoneInfo("America/Sao_Paulo")

"""
Setting constants and variables.
"""
QUANTIFIERS = [10, 100, 500, 1000]

"""
Factorial function
"""
def factorial(n: int):
  if n == 0: # validate if n is 0 or not.
    return 1 # if n is equal 0, return 1.
  rst = n * factorial(n-1) # calc of factorial.
  return rst # return the final value of factorial.

"""
Function breakline
"""
def breakline():
  print("\n")

"""
Loop with requested values
"""
for i in range(len(QUANTIFIERS)):
  print(f"\x1b[36;1mFactorial - With {QUANTIFIERS[i]} itens:\x1b[0m")
  start = dt.datetime.now(tz)
  print(f"\t\x1b[36;1m[Start]⇨ {start}\x1b[0m")

  complexity = 8*QUANTIFIERS[i] # complexity of function.
  print(f"\t\x1b[35;2m[Complexity]⇨ {complexity}\x1b[0m")

  factorial(QUANTIFIERS[i])

  end = dt.datetime.now(tz)
  print(f"\t\x1b[36;1m[End]⇨ {end}\x1b[0m")
  print(f"\t\x1b[36;1m[Interval]⇨ {(end-start).microseconds}ms\x1b[0m")
  breakline()

print(f"\x1b[95;1mThe factorial algorithm with recursive implementation has a complexity of type Linear O(n).",
      "\nThe asymptotic mathematics found was: \'8 * n times\'.",
      "\n\nFunction factorial:",
      "\n\tLine 1 ⇨ 1 access + 1 comparation",
      "\n\tLine 2 ⇨ 1 return",
      "\n\tLine 3 ⇨ 1 assignment + 1 operation + 1 call function + 1 operation",
      "\n\tLine 4 ⇨ 1 return")
```
