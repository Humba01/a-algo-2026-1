Consider the following recursive function:

F(n)=2\*F(n-1)+n\*2

with the base case F(1)=2

Task:
* Implement this function in Python using recursion.
* The program should ask the user for a value of n and calculate F(n).

## **Implementation**

```python
import zoneinfo as zi

"""
Setting the zone info
"""
tz = zi.ZoneInfo("America/Sao_Paulo")

"""
F(n) function
"""
def F(n):
    if n == 1:
        return 2

    return 2 * F(n - 1) + n * 2

n = int(input("Insert one value of n: "))

result = F(n)
print(f"\x1b[34;1mF({n}) = {result}\x1b[0m")
```
