## **Resolution**

### **1. Algorithms**

**i) Merge Sort**

Recurrence: $T(n) = 2T(n/2) + \Theta(n)$

Using the Master Theorem, $a=2, b=2, f(n)=n$. Since $n^{\log_b a} = n^{\log_2 2} = n^1$

It falls under Case 2 where $f(n)$ is equal to the limit of the work of the recursive calls.

Complexity equal to $\Theta(n \log n)$ in all cases.

**ii) Matrix Multiplication**

Recursion: Does not apply (3 loops) by the Master Theorem.

The multiplication of two $n \times n$ matrices: each of the $n^2$ elements of the resulting matrix is the sum of $n$ multiplications.

The calculation involves 3 nested loops (i, j, k), each ranging from $1$ to $n$. Since $n^3 > n^2$.

Therefore, the complexity is $\Theta(n^3)$ in the most common case.


### **2. Recursions**

Using Master's Theorem, we will compare $f(n)$ with $n^{\log_b a}$.

a) $T(n) = 2T(n/4) + \sqrt{n}$
- $a = 2, b = 4, f(n) = n^{1/2}$
- Calculate $n^{\log_b a} = n^{\log_4 2} = n^{1/2}$
- The result found was $f(n) = \Theta(n^{1/2})$, so we fall into Case 2.
- The complexity found was $T(n) = \Theta(\sqrt{n} \log n)$

b) $T(n) = 2T(n/4) + n$
- $a = 2, b = 4, f(n) = n^1$
- Calculate $n^{\log_4 2} = n^{1/2}$
- Comparison showing that $f(n) = n^1$ grows faster than $n^{1/2}$.
- We fall into Case 3 (the term outside the recursion dominates).
- The complexity found was $T(n) = \Theta(n)$

c) $T(n) = 16T(n/4) + n^2$
- $a = 16, b = 4, f(n) = n^2$
- Calculate $n^{\log_4 16} = n^2$
- The result found was Case 2, where the division work and the combination work are equivalent.
- The complexity found was $T(n) = \Theta(n^2 \log n)$
