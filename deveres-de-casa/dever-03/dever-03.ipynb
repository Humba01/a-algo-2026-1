```python
import datetime as dt
import numpy as np
import os
import random as rnd
import zoneinfo as zi
import array as arr

"""
Setting the zone info
"""
tz = zi.ZoneInfo("America/Sao_Paulo")

"""
Palindrome implementation with definition of array before.
"""
myarr = ['a', 'b', 'c', 'c', 'b', 'a']

def palindrome(list: arr):
  if len(list) % 2 != 0:
    print("\x1b[31;1mIsn't a palindrome!\x1b[0m")
  else:
    tam_total = int(len(list))
    tam_metade = int(len(list)/2)
    normal = list[0:tam_metade]
    reverse = list[tam_metade:tam_total]
    reverse.reverse()
    if normal == reverse:
      print("\x1b[34;1mIs a palindrome!\x1b[0m")
    else:
      print("\x1b[31;1mIsn't a palindrome!\x1b[0m")

palindrome(myarr)
```
